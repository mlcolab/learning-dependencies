# Annotations processing
The goal is to decide on relevant annotations and filter them for mathematical concepts.

```python
import pickle
import os
from matplotlib import pyplot as plt
from itertools import chain
```

```python
path = "../dat/annotations/"
books = list(os.listdir(path))

annos = {}
for book in books:
    print(book)
    annos[book] = pickle.load(open(path + book, "rb" ))
    print(len(annos[book]))
```

```python
annos[books[0]][1000]
```

```python
prs = []
for b in annos.values():
    prs.append([[s['pageRank'] for s in a['support']] for a in b])
```

```python
prs_flat = list(chain.from_iterable(prs))
```

```python
len(prs_flat)
```

```python
plt.hist(prs_flat, bins=10)
#plt.xlim(0, 0.001)
plt.show()
```

```python
a_dict = {}
not_math_concepts = set()

for a in annos:
    for s in a['support']:
        if s['pageRank'] > 0.0001:
            if a['title'] not in a_dict:
                a_dict[a['title']] = []
            a_dict[a['title']].append(s['chFrom'])

    #if 'wikiDataClasses' in a and 'mathematical concept' in [c['enLabel'] for c in a['wikiDataClasses']]: 
    #else:
    #    not_math_concepts.add(a['url'])
```
